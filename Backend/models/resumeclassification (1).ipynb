```python
import pandas as pd
import re
import spacy
import joblib
import os

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sentence_transformers import SentenceTransformer

# ===============================
# Step 1: Load and clean dataset
# ===============================

# Load your dataset
df = pd.read_csv("final_combined_resumes.csv")

# Normalize label names
df['label'] = df['label'].str.lower().str.strip().replace({
    'data science': 'data scientist',
    'datascience': 'data scientist',
    'web designing': 'web developer',
    'webdesigning': 'web developer',
})

# Drop classes with fewer than 2 samples (important for stratified split)
value_counts = df['label'].value_counts()
valid_labels = value_counts[value_counts >= 2].index
df = df[df['label'].isin(valid_labels)]

# ===============================
# Step 2: Clean resume text
# ===============================

# Load spaCy and define stopwords
nlp = spacy.load("en_core_web_sm")
stopwords = nlp.Defaults.stop_words

# Text preprocessing function
def clean_text(text):
    text = re.sub(r'\S+@\S+', '', text)         # Remove email addresses
    text = re.sub(r'\b\d{10,}\b', '', text)     # Remove long numbers (e.g., phone numbers)
    text = re.sub(r'\b\d+\b', '', text)         # Remove all numbers
    text = re.sub(r'[^\w\s]', '', text)         # Remove punctuation
    text = text.lower()

    doc = nlp(text)
    tokens = [token.lemma_ for token in doc
              if token.text not in stopwords and not token.is_stop and not token.is_punct]

    return " ".join(tokens)

# Apply cleaning
df['clean_text'] = df['resume_text'].astype(str).apply(clean_text)

# ===============================
# Step 3: Embedding + Label Encoding
# ===============================

# Load Sentence-BERT model
embedder = SentenceTransformer('all-MiniLM-L6-v2')

# Embed resume texts
X = embedder.encode(df['clean_text'].tolist(), show_progress_bar=True)

# Encode target labels
le = LabelEncoder()
y = le.fit_transform(df['label'])

# ===============================
# Step 4: Train-test split
# ===============================

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

# ===============================
# Step 5: Define models and grid
# ===============================

models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "SVM": SVC(),
    "RandomForest": RandomForestClassifier()
}

param_grid = {
    "LogisticRegression": {"C": [0.1, 1, 10]},
    "SVM": {"C": [0.1, 1, 10], "kernel": ["linear", "rbf"]},
    "RandomForest": {"n_estimators": [50, 100], "max_depth": [None, 10]}
}

# ===============================
# Step 6: Grid Search and Best Model Selection
# ===============================

best_model = None
best_score = 0
best_name = ""
best_params = {}

for name, model in models.items():
    print(f"\n🔍 Running GridSearchCV for {name}")
    grid = GridSearchCV(model, param_grid[name], cv=3, scoring='f1_weighted', n_jobs=-1)
    grid.fit(X_train, y_train)

    score = grid.best_score_
    print(f"✅ Best F1 Score: {score:.4f}")
    print(f"✅ Best Params: {grid.best_params_}")

    if score > best_score:
        best_score = score
        best_model = grid.best_estimator_
        best_name = name
        best_params = grid.best_params_

# ===============================
# Step 7: Evaluate and Save
# ===============================

print(f"\n🏆 Best Model Selected: {best_name}")
y_pred = best_model.predict(X_test)
print("\n📊 Classification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# Save the model and label encoder
os.makedirs("exported_model", exist_ok=True)
joblib.dump(best_model, "exported_model/resume_classifier.pkl")
joblib.dump(le, "exported_model/label_encoder.pkl")

print("\n✅ Model and Label Encoder saved to 'exported_model/' folder.")
```
